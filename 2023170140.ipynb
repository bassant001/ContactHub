#building linear regression

```python
def linear_regression(x,y,l,epochs):
  m=0
  c=0
  n = float(len(x)) # number of elements in X

  for i in range(epochs):
    yp= m * x + c

    m = m - l * (- (1/n) * np.sum((y - yp) * x))
    c = c - l * (- (1/n) * np.sum(y - yp))

  return (m, c)
```

#building multiple regration

```python
def multiple_regration(x1,x2,y,l,epochs):
  m1=0
  m2=0
  c=0
  n = float(len(x1)) # number of elements in X

  for i in range(epochs):
    yp = (m1 * x1) + (m2 * (x2)) + (c)

    m1 = m1 - l * (-(1/n) * np.sum((y - yp) * x1))
    m2 = m2 - l * (-(1/n) * np.sum((y - yp) * (x2)))
    c = c - l * (-(1/n) * np.sum(y - yp))

  return (m1,m2, c)
```

#features

```python
x1 = data["Square Footage"].values
x2 = data["Number of Occupants"].values
x3 = data["Appliances Used"].values
x4 = data["Average Temperature"].values
```

#target

```python
y  = data["Energy Consumption"].values
```

#The learning Rate
#The number of iterations to perform gradient descent

```python
learning_rates = [1e-12,1e-11,1e-10,1e-9]
epochs = [100 , 200 , 300 , 400]
```

#apply linear regression model



```python
X =[x1, x2, x3, x4] #features

for i in range(len(X)):
  x = X[i]
  print("X",i," :")
  print("#######################")

  for l in range(len(learning_rates)):
    for e in range(len(epochs)):
      m,c = linear_regression(x,y,learning_rates[l],epochs[e])
      yp = m * x + c
      mse = np.mean((y - yp)**2)
      rmse = np.sqrt(mse)
      mape = np.mean(np.abs((y - yp) / y)) * 100

      print("MSE:", mse)
      print("RMSE:", rmse)
      print("MAPE:", mape)
      print("Learning Rate: ",learning_rates[l])
      print("Epochs: ",epochs[e])
      print("Slope: ",m)
      print("Intercept: ",c)
      print("___________________________________________")
  print("\n\n")

```

#apply multiple regression model


```python
x =[x1, x2, x3, x4] #features

for i in range(2):
  print("X",i," :", "X",i+1," :")
  print("#######################")

  for l in range(len(learning_rates)):
    for e in range(len(epochs)):
      m1,m2,c = multiple_regration(x[i],x[i+1],y,learning_rates[l],epochs[e])
      yp = (m1 * x[i]) + (m2 * (x[i+1])) + (c)
      mse = np.mean((y - yp)**2)
      rmse = np.sqrt(mse)
      mape = np.mean(np.abs((y - yp) / y)) * 100

      print("MSE:", mse)
      print("RMSE:", rmse)
      print("MAPE:", mape)
      print("Learning Rate: ",learning_rates[l])
      print("Epochs: ",epochs[e])
      print("Slope: ",m)
      print("Intercept: ",c)
      print("___________________________________________")
  print("\n\n")
```
